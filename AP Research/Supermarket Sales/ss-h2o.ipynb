```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Initialize H2O
h2o.init()
```

```python
# Load the dataset (replace this with your actual dataset loading code)
df = pd.read_csv('/content/drive/MyDrive/ML/AP Research/supermarket_sales - Sheet1.csv')

# Clean column names (if needed)
df.columns = df.columns.str.replace('[^a-zA-Z0-9_]', '_', regex=True)

# Convert the pandas DataFrame to an H2O Frame
h2o_df = h2o.H2OFrame(df)

# Define the target column and feature columns
target = "Rating"
features = h2o_df.columns
features.remove(target)  # Remove the target column from the feature list

# Split the data into training and testing sets
train, test = h2o_df.split_frame(ratios=[0.8], seed=42)
```

```python
# Initialize H2O AutoML
aml = H2OAutoML(max_models=10, seed=42, max_runtime_secs=900)  # 15 minutes runtime

# Train the model
aml.train(x=features, y=target, training_frame=train)
```

```python
# Get the best model from AutoML
best_model = aml.leader

# Make predictions on the test set
preds = best_model.predict(test)

# Convert H2O Frame to pandas DataFrame for evaluation
y_test = test[target].as_data_frame()
y_pred = preds.as_data_frame()

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)

# Calculate R-squared (R²)
r2 = r2_score(y_test, y_pred)
print("R-squared (R²):", r2)
```

```python
# View the leaderboard
lb = aml.leaderboard
print(lb)
```
